Notebook containing the preprocessing steps for the steel.csv dataset for A2.

```python
#import steel dataset and store 

import pandas as pd

pd.set_option('display.width', 200)
df = pd.read_csv('steel.csv')

print("pandas dataframe:\n", df.head())

tf = df.to_numpy()
print("\nnumpy array:\n", tf[:5])
```

The Features (X) and targets (y) should be separated.

```python
X_features = tf[:, 0:9]
print("\nFeatures (X):\n", X_features[:5])

y_target = tf[:, 9]
print("\nTarget (y):\n", y_target[:5])
```

The data must be split into training and testing datasets as a part of the preprocessing using test_train_split from sklearn.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.2, random_state=42)

# store split data for use in other notebooks
%store X_train
%store X_test
%store y_train
%store y_test
```
